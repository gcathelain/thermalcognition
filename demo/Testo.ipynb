# Method 1 : Colorbar

The colorbar of each image, whose minimum and maximum values are automatically detected, is used to infer the temperature values of each image.

```python
video_path = r"../data/2020_01_23_Coralynn.mpg"
```

```python
video_colorbar = read_testo_video(video_path)
```

```python
frame0 = 0
frame1 = 30000
plt.subplot(121)
plt.imshow(video_colorbar[frame0].astype(np.float64), cmap="gray")
plt.colorbar()
plt.title("Temperature of frame {0} with colorbar method".format(frame0))
plt.subplot(122)
plt.imshow(video_colorbar[frame1].astype(np.float64), cmap="gray")
plt.title("Temperature of frame {0} with colorbar method".format(frame1))
plt.colorbar()
```
